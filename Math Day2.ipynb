```python
# converting given integer into fraction and in percentage

user_input = input("Enter a number")

#find number of decimal places
exp = int(len(user_input))-1

n = float(user_input)

numerator = int(n * 10 ** exp)
denominator = 10**exp

percent = n * 100

print("fraction is:",numerator,"/",denominator)
print("Percentge value is:",percent)
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
x_min = -10
x_max = 10
y_min = -10
y_max = 10
plt.axis([x_min,x_max,y_min,y_max])

plt.plot([x_min,x_max],[0,0])
plt.plot([0,0],[y_min,y_max])

for x in range(x_min,x_max+1):
    y = x+5
    plt.plot([x],[y],'ro')
plt.show()
```

```python
# Rather than using loop lets use numpy array

import matplotlib.pyplot as plt
import numpy as np

x_min = -10
x_max = 10
y_min = -10
y_max = 10

# Number of points needed
points = 2*(x_max-x_min)
x = np.linspace(x_min,x_max,points)
fig,ax = plt.subplots()
plt.axis([x_min,x_max,y_min,y_max])


ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.grid(True)

y=2*x+1
plt.plot(x,y,'r',label="y=2*x+1")
plt.plot([x_min,x_max],[0,0],'g',label="y=0")
plt.plot([0,0],[y_min,y_max],label="x=0")
plt.legend()
plt.show()
```

```python
#Factoring

digits = input("Enter a decimal number to covert:")

#convert to fraction
exponent = int(len(digits))-1
n = float(digits)

numerator = int(n*10**exponent)
denominator = 10**exponent

# reduce the fraction
factor = 1
for test_factor in range(1,denominator+1):
    if numerator%test_factor==0 and denominator%test_factor==0:
        factor = test_factor
        
num = int(numerator/factor)
den = int(denominator/factor)

print(n)
print(num,'/',den)
```
