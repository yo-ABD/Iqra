```python
# when you try to print fraction when taking it as input you get error
#x = float(input())
#print(x)
# to solve that issue we can use the below function


# Convert string inputs to float:
def string_frac(in_string):
    if "/" in in_string:
        #splitting to get numerator and denominator
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
    else:
        ans = float(in_string)
    
    return ans
```
